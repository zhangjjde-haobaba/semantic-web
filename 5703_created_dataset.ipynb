```python
from faker import Faker
import random

# 创建Faker对象
fake = Faker()

# Initialize variables
num_projects = 10  # Number of projects
roles_per_project = 4  # Number of roles per project

# 生成随机数据
data = []

# for _ in range(100):
#     first_name = fake.first_name()
#     last_name = fake.last_name()
#     country = fake.country()
#     languages = [fake.language_name() for _ in range(random.randint(1, 5))]  # 随机生成1到5种语言

#     data.append({
#         'First Name': first_name,
#         'Last Name': last_name,
#         'Country Location': country,
#         'Languages': ', '.join(languages)  # 将语言列表转换为逗号分隔的字符串
#     })

# Generate data
num_projects = 5  # Number of projects
roles_per_project = 4  # Number of roles per project

for _ in range(num_projects):
    project_name = fake.company_suffix()
    project_id = fake.random_int(min=10000, max=99999)

    for _ in range(roles_per_project):
        role_name = fake.random_element(available_roles)
        role_description = "This is a role for " + role_name
        hourly_rate = fake.random_number(digits=2)
        required_skills = ", ".join(fake.random_elements(skills_vocabulary_full, length=3))
        preferred_skills = ", ".join(fake.random_elements(skills_vocabulary_full, length=2))
        tools = ", ".join(fake.random_elements(available_tools, length=3))
        time_zone = fake.timezone()
        availability_date = fake.date_this_month()
        min_hours_per_week = fake.random_int(min=10, max=40)
        location = fake.city()
        years_of_experience = fake.random_int(min=1, max=10)
        language = fake.random_element(available_languages)
        about_me = fake.sentence()

        data.append([
            project_name, project_id, role_name, role_description, hourly_rate,
            required_skills, preferred_skills, tools, time_zone, availability_date,
            min_hours_per_week, location, years_of_experience, language, about_me
        ])

# Create a DataFrame with updated data
df_updated = pd.DataFrame(data, columns=[
    'Project Name', 'Project ID', 'Role Name', 'Role Description', 'Hourly Rate',
    'Required Skills', 'Preferred Skills', 'Tools', 'Time Zone/Available Hours',
    'Availability Date', 'Minimum Available Hours per Week', 'Location',
    'Years of Experience', 'Language', 'About Me'
])
```

```python
# Write the generated data to an Excel file in a sheet named 'roles'
excel_file_path = 'Book2.xlsx'
df_updated.to_excel(excel_file_path, sheet_name='roles', index=False)
```

```python
# Initialize list to hold generated data
talent_data_list = []

# Number of talents to generate
num_talents = 100

# Generate data for each talent
for _ in range(num_talents):
    first_name = fake.first_name()
    last_name = fake.last_name()
    country_location = fake.country()
    availability_date = fake.date_this_month()
    availability_hours_pw = fake.random_int(min=10, max=40)
    role = fake.random_element(available_roles)
    years_of_experience = fake.random_int(min=1, max=10)
    skills = ", ".join(fake.random_elements(skills_vocabulary_full, length=3))
    social_links = fake.url()
    short_biography = fake.paragraph()
    personality_questions = ", ".join(fake.random_elements(available_personality_questions, length=3))
    tools = ", ".join(fake.random_elements(available_tools, length=3))
    languages = ", ".join(fake.random_elements(available_languages, length=2))
    work_experience_title = fake.job()
    work_experience_company = fake.company()
    work_experience_years = fake.random_int(min=1, max=5)
    work_experience_industry = fake.random_element(['IT', 'Finance', 'Healthcare', 'Education'])
    projects = fake.company_suffix()  # Placeholder, will replace with actual project names
    hourly_rate = fake.random_number(digits=2)
    working_hours = fake.random_int(min=20, max=40)

    talent_data_list.append([
        first_name, last_name, country_location, f"From {availability_date}, {availability_hours_pw} hours/pw",
        role, years_of_experience, skills, social_links, short_biography, personality_questions,
        tools, languages, f"{work_experience_title} at {work_experience_company}, {work_experience_years} years in {work_experience_industry}",
        projects, hourly_rate, working_hours
    ])

# Create a DataFrame with generated talent data
df_talent = pd.DataFrame(talent_data_list, columns=[
    'First Name', 'Last Name', 'Country Location', 'Availability (Available from date, hours p/w)',
    'Role', 'Years of Professional Experience', 'Skills', 'Social Links', 'Short Biography',
    '3 x Personality Questions', 'Tools', 'Languages', 'Work Experience (Title, Company, Years, Industry)',
    'Projects', 'Hourly Rate', 'Working Hours'
])

# Display the first few rows of the generated DataFrame
df_talent.head()
```
